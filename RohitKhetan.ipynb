```python
#Identifying and dropping the columns that are irrelevant to the analysis.
data_loan.drop(['id', 'member_id', 'title', 'emp_title','url', 'zip_code', 'addr_state', 'delinq_2yrs','total_pymnt','total_pymnt_inv','total_rec_prncp','total_rec_int','total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','last_credit_pull_d', 'out_prncp','out_prncp_inv', 'pub_rec_bankruptcies','revol_util' ], inplace=True, axis = 1)
```

```python
#Handling missing values 
```

```python
#Dropping the columns where count of null values = the total number of rows
rows = data_loan.shape[0]
for x in data_loan.columns:
    if data_loan[x].isnull().sum()==rows:
        data_loan.drop([x],inplace=True, axis=1)
```

```python
#Checking the shape again
data_loan.shape
```

Finding: Columns with all null values are dropped from the analysis

```python
#Checking the number of null values in the columns:
data_loan.isnull().sum()
```

Finding: There are still multiple columns with null. Upon analysis the columns with more than 2500 null values are considered irrelevant to our analysis, hence can be dropped.

```python
#Dropping the columns where count of null values >2500, as such columns don't add value to the analysis. 
##Upon further review these columns are irrelevant to the analysis.
for x in data_loan.columns:
    if data_loan[x].isnull().sum()>2500:
        data_loan.drop([x],inplace=True, axis=1)
```

```python
#Checking the shape again
data_loan.shape
```

Finding: Intended coloumns are dropped from the data set.

```python
#Checking the number of unique variables per column:
data_loan.nunique()
```

Finding: There are multiple columns with single unique value and irrelevant to our analysis.

```python
#Dropping the columns where count of unique values is equal to 1, as such columns don't add value to the analysis. 
##Upon further review these columns are irrelevant to the analysis.
for x in data_loan.columns:
    if data_loan[x].nunique()==1:
        data_loan.drop([x],inplace=True, axis=1)
```

```python
#Checking the number of unique variables per column:
data_loan.nunique()
```
